```python
import pandas as pd

import matplotlib.pyplot as plt

import lightgbm as lgb
from lightgbm import LGBMModel,LGBMClassifier

from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
# read in data
train = pd.read_csv("dataset/TrainingSet.csv")
train_labels = pd.read_csv("dataset/TrainingSetLabels.csv")
test = pd.read_csv("dataset/TestSet.csv")
```

```python
# find categorical columns

categorical = [name for t, name in zip(train.dtypes, list(train)) if t == object]
```

```python
# remove pump id

easy_train = train.drop(["id"], axis=1)
easy_sub = test.drop(["id"], axis=1)
train_labels = train_labels.drop(["id"], axis=1)
```

```python
# transform categorical features into appropriate types

for col in easy_train.columns:
    if col in categorical:
        easy_train[col] = easy_train[col].astype("category")
        easy_sub[col] = easy_sub[col].astype("category")
```

```python
# set aside test set

X_train, X_test, y_train, y_test = train_test_split(easy_train, train_labels, test_size=0.15, random_state=42)
```

```python
# lgbm

# creat lgbm classfier

lgb = LGBMClassifier(n_estimators=1000, 
                    silent=False, 
                    random_state=43,
                    categoricalfeature=categorical
                     )
```

```python
# fit with train data 

lgb.fit(X_train, y_train)
```

```python
imps = lgb.feature_importances_
```

```python
# plot feature importance

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel("Importance")
ax.set_ylabel("Feature Label")
ax.set_title("Feature Importances")

ax.barh(list(X_train), imps)
```

```python
# predict 
pred = lgb.predict(X_test)
```

```python
print(metrics.classification_report(pred, y_test))
```

```python
# predict with lgb

lgb_prediction = lgb.predict(easy_sub)
```
